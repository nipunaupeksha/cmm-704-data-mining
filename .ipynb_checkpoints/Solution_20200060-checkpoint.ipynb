***1.1.3.2 Job and Education***

Now, to infer the missing values in ***'job' and 'education'***, we make use of the cross-tabulation between 'job' and 'education'. We assume that in here 'job' is influenced by the 'education' of a person. Hence, we can infer 'job' based on the education of the person. Moreover, since we are just filling the missing values, we are not much concerned about the causal inference. We, therefore, can use the job to predict the education.

```python
# Cross tabulation between both two columns
def cross_tab(data, col_name_1 , col_name_2):
    # Find number of unique values in firdy column
    col_one = list(data[col_name_1].unique())
    # Find no of unique values in second columns
    col_two = list(data[col_name_2].unique())
    dataframe_list = []
    for col in col_two:
        column_one_dataframe = data[data[col_name_2] == col]
        both_column_dataframe = column_one_dataframe.groupby(col_name_1).count()[col_name_2]
        dataframe_list.append(both_column_dataframe)
    cross_tab_res = pd.concat(dataframe_list,axis=1)
    cross_tab_res.columns = col_two

    # Fill null values with zero (0)
    cross_tab_res = cross_tab_res.fillna(0)
    return cross_tab_res
```

```python
# Cross tabulation between job and education
cross_tab(banking_df_missing_value_fix,'job','education')
```

* Inferring education from jobs : From the cross-tabulation, it can be seen that people with management jobs usually have a university degree. Hence wherever 'job' = management and 'education' = unknown, we can replace 'education' with 'university.degree'. Similarly, 
  * 'job' = 'services' -> 'education' = 'high.school' 
  * 'job' = 'housemaid' -> 'education' = 'basic.4y'.

* Inferring jobs from education : If 'education' = 'basic.4y' or 'basic.6y' or 'basic.9y' then the 'job' is usually 'blue-collar'. If 'education' = 'professional.course', then the 'job' = 'technician'.

* While impuding the values for job and education, we were cognizant of the fact that the correlations should make real world sense. If it didn't make real world sense, we don't need to replace the missing values.

```python
# Replace the Unknown Values
banking_df_missing_value_fix.loc[(banking_df_missing_value_fix['education']=='unknown') & (banking_df_missing_value_fix['job']=='management'), 'education'] = 'university.degree'
banking_df_missing_value_fix.loc[(banking_df_missing_value_fix['education']=='unknown') & (banking_df_missing_value_fix['job']=='services'), 'education'] = 'high.school'
banking_df_missing_value_fix.loc[(banking_df_missing_value_fix['education']=='unknown') & (banking_df_missing_value_fix['job']=='housemaid'), 'education'] = 'basic.4y'
banking_df_missing_value_fix.loc[(banking_df_missing_value_fix['job'] == 'unknown') & (banking_df_missing_value_fix['education']=='basic.4y'), 'job'] = 'blue-collar'
banking_df_missing_value_fix.loc[(banking_df_missing_value_fix['job'] == 'unknown') & (banking_df_missing_value_fix['education']=='basic.6y'), 'job'] = 'blue-collar'
banking_df_missing_value_fix.loc[(banking_df_missing_value_fix['job'] == 'unknown') & (banking_df_missing_value_fix['education']=='basic.9y'), 'job'] = 'blue-collar'
banking_df_missing_value_fix.loc[(banking_df_missing_value_fix['job']=='unknown') & (banking_df_missing_value_fix['education']=='professional.course'), 'job'] = 'technician'

cross_tab(banking_df_missing_value_fix,'job','education')
```

Now we can see the no of unknown values are reduced and data set is some what optimized

***1.1.3.3 House and Job***<br>
We are again using cross-tabulation between ***'house' and 'job'*** Our hypothesis is that housing loan status (Yes or No) should be in the proportion of each job category. Hence using the prior known distribution of the housing loan for each job category, the house loan for unknown people will be predicted such that the prior distribution (% House = Yes's and No's for each job category remains the same).

```python
job_housing_df = cross_tab(banking_df_missing_value_fix,'job','housing')
print(job_housing_df)
```

```python
# Function for imputation via cross-tabulation to fill missing values for the 'housing' categorical feature
def fillhousing(data,jobhousing):
    jobs = ['housemaid','services','admin.','blue-collar','technician','retired','management','unemployed','self-employed','entrepreneur','student']
    house = ['no','yes']
    for j in jobs:
        # Here we are taking the value where housing is unknown and job value is known
        ind = data[np.logical_and(np.array(data['housing']=='unknown'),np.array(data['job']==j))].index
        mask = np.random.rand(len(ind))<((jobhousing.loc[j]['no'])/(jobhousing.loc[j]['no']+jobhousing.loc[j]['yes']))
        ind1 = ind[mask]
        ind2 = ind[~mask]
        data.loc[ind1,'housing'] = 'no'
        data.loc[ind2,'housing'] = 'yes'
    return data

banking_df_missing_value_fix=fillhousing(banking_df_missing_value_fix, job_housing_df)

jobhousing=cross_tab(banking_df_missing_value_fix,'job','housing')
print(jobhousing)
```

***1.1.3.4 Loan and Job***<br>
We are again using cross-tabulation between ***'house' and 'job'*** Our hypothesis is that loan status (Yes or No) should be in the proportion of each job category.
```python
# Function for imputation via cross-tabulation to fill missing values for the 'loan' categorical feature
def fillloan(data,job_loan_df):
    jobs = ['housemaid','services','admin.','blue-collar','technician','retired','management','unemployed','self-employed','entrepreneur','student']
    loan = ['no','yes']
    for j in jobs:
        ind = data[np.logical_and(np.array(data['loan']=='unknown'),np.array(data['job']==j))].index
        mask = np.random.rand(len(ind))<((job_loan_df.loc[j]['no'])/(job_loan_df.loc[j]['no']+job_loan_df.loc[j]['yes']))
        ind1 = ind[mask]
        ind2 = ind[~mask]
        data.loc[ind1,'loan']='no'
        data.loc[ind2,'loan']='yes'
    return banking_df_missing_value_fix

banking_df_missing_value_fix = fillloan(banking_df_missing_value_fix,job_loan_df)

job_loan_df = cross_tab(banking_df_missing_value_fix,'job','loan')
print(job_loan_df)
```

```python
banking_df_missing_value_fix.head()
```

```python
# Check quantiles before handeling outliers
def checkQuantiles(df):
  for column in num_fields:
      print(f'For {column}:')

      print('Min:', df[column].quantile(q = 0))
      print('25% Quartile:', df[column].quantile(q = 0.25))
      print('50% Quartile:', df[column].quantile(q = 0.50))
      print('75% Quartile:', df[column].quantile(q = 0.75))
      print('Max:', df[column].quantile(q = 1.00),'\n')

checkQuantiles(banking_df_missing_value_fix)
```

```python
# Handling outliers using Quantile Method
# new outlier removed data set
banking_df_outliers1 = banking_df_missing_value_fix.copy();

def outliersUsingQuantile(field):
  fig, axes = plt.subplots(1,2)
  plt.tight_layout()
  print("Before Shape:",banking_df_missing_value_fix.shape);
  ## Max and Min Quantile
  max_val = banking_df_missing_value_fix[field].quantile(0.95);
  min_val = banking_df_missing_value_fix[field].quantile(0.05);
  ## Removing all the outliers
  banking_df_outliers1 = banking_df_missing_value_fix[(banking_df_missing_value_fix[field]>min_val) & (banking_df_missing_value_fix[field]<max_val)];
  ## Visulization
  print("After Shape:",banking_df_outliers1.shape);
  sns.boxplot(banking_df_missing_value_fix[field],orient='v',ax=axes[0]);
  axes[0].title.set_text("Before");
  sns.boxplot(banking_df_outliers1[field],orient='v',ax=axes[1]);
  axes[1].title.set_text("After");
  plt.show();

for field in num_fields :
  outliersUsingQuantile(field)
```

```python
banking_df_outliers1.shape
```

```python
checkQuantiles(banking_df_outliers1)
```

```python
# Handling outliers using IQR method
# new outlier removed data set
banking_df_outliers2 = banking_df_missing_value_fix.copy();

def outliersUsingIQR(field):
  fig, axes = plt.subplots(1,2)
  plt.tight_layout()
  print("Previous Shape With Outlier: ",banking_df_missing_value_fix.shape)
  sns.boxplot(banking_df_missing_value_fix[field],orient='v',ax=axes[0])
  axes[0].title.set_text("Before")

  # claculate IQR
  Q1 = banking_df_missing_value_fix[field].quantile(0.25)
  Q3 = banking_df_missing_value_fix[field].quantile(0.75)
  print('Q1 :', Q1, ' Q3 :', Q3)
  IQR = Q3-Q1
  print('IQR : ', IQR)
  lower_limit = Q1 - 1.5*IQR
  upper_limit = Q3 + 1.5*IQR
  print('Lower limit :', lower_limit, ' Uppre limit :', upper_limit)
  banking_df_outliers2[field] = np.where(banking_df_outliers2[field] > upper_limit,upper_limit,banking_df_outliers2[field])
  banking_df_outliers2[field] = np.where(banking_df_outliers2[field]<lower_limit,lower_limit,banking_df_outliers2[field])
  print("Shape After Removing Outliers:", banking_df_outliers2.shape)
  sns.boxplot(banking_df_outliers2[field],orient='v',ax=axes[1])
  axes[1].title.set_text("After")
  plt.show()

for field in num_fields :
  outliersUsingIQR(field)
```

### ***1.2 SVD (Singular Value Decomposition)***




```python
from scipy.sparse import csr_matrix
from sklearn.decomposition import TruncatedSVD

df_for_svd_tmp = banking_df_discretization.copy()

# Make sparse matrix
X_sparse = csr_matrix(df_for_svd_tmp)
X_sparse.shape[1]-1

tsvd = TruncatedSVD(n_components=X_sparse.shape[1]-1)
X_tsvd = tsvd.fit(df_for_svd_tmp)

tsvd_var_ratios = tsvd.explained_variance_ratio_

def select_n_components(var_ratio, goal_var: float) -> int:
    # Set initial variance explained so far
    total_variance = 0.0
    
    # Set initial number of features
    n_components = 0
    
    # For the explained variance of each feature:
    for explained_variance in var_ratio:
        # Add the explained variance to the total
        total_variance += explained_variance
        n_components += 1
        
        # If we reach our goal level of explained variance
        if total_variance >= goal_var:
            break
            
    return n_components

number_of_components = select_n_components(tsvd_var_ratios, 0.95)
print("Number of components : ", number_of_components)
```

```python
## SVD

from sklearn.decomposition import TruncatedSVD

pca_categorical = smoted_X.copy().drop(numerical_variables,axis=1)

df_for_svd = pd.concat([banking_df_discretization, pca_categorical],join="outer",axis=1)

svd = TruncatedSVD(n_components = number_of_components)
#svd = TruncatedSVD()
# prepare transform on dataset
svd.fit(df_for_svd)

# apply transform to dataset
transformed = svd.transform(df_for_svd)

df_svd = pd.DataFrame(transformed)
df_svd.head()
```

# **(c) SVM and LR**

```python
# dividing the dataset as training and testing
from sklearn.model_selection import train_test_split

processed_data = df_pca.copy()

# Examine the shape of the DataFrame (again)
print(processed_data.shape)

X = processed_data
# Saperating features and result vectors

X_train, X_test, y_train, y_test = train_test_split(X, data_y, test_size=0.20, random_state=42)

print("Train dataset:", X_train.shape)
print("Test dataset:", X_test.shape)
```

#### **1. SVM**

```python
#Import svm model
from sklearn import svm
from sklearn import metrics

# Create a svm Classifier with Linear Kernel
classifier_svm = svm.SVC(kernel='linear') 

# Train the model using the training sets
classifier_svm.fit(X_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix
import math

# Predict the response for test dataset
y_pred = classifier_svm.predict(X_test)

# evaluation
print("Classifier score : ", classifier_svm.score(X_test, y_test)) 
print("Confusion matrix : ", confusion_matrix(y_test,classifier_svm.predict(X_test)))

# accuracy testing
accuracy_svm = metrics.accuracy_score(y_test, y_pred)
print("Accuracy score : ", accuracy_svm)
precision_svm = metrics.precision_score(y_test, y_pred)
print("Precision score : ", precision_svm)
recall_svm = metrics.recall_score(y_test, y_pred)
print("recall_score : ", recall_svm)
f1_svm = 2 * precision_svm * recall_svm / (precision_svm + recall_svm)
print("f1 score : ", f1_svm)
```

#### **2. Logistic Regression**

```python
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)

classifier_lr = LogisticRegression(random_state=0)

classifier_lr.fit(X_train,y_train) 
```

```python
# evaluation
print("Classifier score : ", classifier_lr.score(X_test,y_test)) 
print("Confusion matrix : ", confusion_matrix(y_test,classifier_lr.predict(X_test)))

# accuracy testing
accuracy_lr = accuracy_score(y_test,classifier_lr.predict(X_test))
print("Accuracy score : ", accuracy_lr)
precision_lr = precision_score(y_test,classifier_lr.predict(X_test))
print("Precision score : ", precision_lr)
recall_lr = recall_score(y_test,classifier_lr.predict(X_test))
print("recall_score : ", recall_lr)
f1_lr = 2 * precision_lr * recall_lr / (precision_lr + recall_lr)
print("f1 score : ", f1_lr)
```

```python
def train_and_evaluate(clf, X_train, X_test, y_train, y_test):
    clf.fit(X_train, y_train)
    print ("Accuracy on training set:")
    print (clf.score(X_train, y_train))
    print ("Accuracy on testing set:")
    print (clf.score(X_test, y_test))
    y_pred = clf.predict(X_test)
    print ("Classification Report:")
    print (metrics.classification_report(y_test, y_pred))
    print ("Confusion Matrix:")
    print (metrics.confusion_matrix(y_test, y_pred))
```
